## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_data = [
    'Age',
    'DistanceFromHome',
    'Education',
    'JobSatisfaction',
    'NumCompaniesWorked',
    'OverTime',
    'StockOptionLevel',
    'WorkLifeBalance',
    'YearsAtCompany',
    'YearsSinceLastPromotion'
]

# Create X_df using your selected columns
X_df = attrition_df[X_data].set_index(attrition_df.index)

# Show the data types for X_df
X_df.info()
```

```python
y_df.set_index(X_df.index)
y_df.head()
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_df=pd.get_dummies(X_df)

display(X_df.head(), X_df.shape)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=26)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
# X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
dept_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
dept_y_train_enc=dept_encoder.fit_transform(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
# dept_y_train_enc = dept_encoder.transform(y_train[['Department']])
dept_y_test_enc = dept_encoder.transform(y_test[['Department']])
```

```python
# Create a OneHotEncoder for the Attrition column
attr_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
# attr_encoder.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
attr_y_train_enc = attr_encoder.fit_transform(y_train[['Attrition']])
attr_y_test_enc = attr_encoder.transform(y_test[['Attrition']])
```

```python
print(dept_y_train_enc.shape)
print(dept_y_test_enc.shape)
print(attr_y_train_enc.shape)
print(attr_y_test_enc.shape)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
columns = X_train_scaled.shape[1]
print(columns)
# Create the input layer
input = layers.Input(shape=((columns), ), name='input')

# Create at least two shared layers
shared_layer_1 = layers.Dense(128, activation = "relu", name='shared_layer_1')(input)
shared_layer_2 = layers.Dense(64, activation = "relu", name='shared_layer_2')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_input = layers.Dense(32, activation="relu")(shared_layer_2)

# Create the output layer
dept_output = layers.Dense(dept_y_train_enc.shape[1], activation='softmax', name='dept_output')(dept_input)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_input = layers.Dense(32, activation="relu")(shared_layer_2)

# Create the output layer
attr_output = layers.Dense(attr_y_train_enc.shape[1], activation='sigmoid', name='attr_output')(attr_input)
```

```python
# Create the model
model = Model(inputs=input, outputs=[dept_output, attr_output])

# Compile the model
model.compile(optimizer='adam', loss={
        'dept_output' : 'categorical_crossentropy',
        'attr_output' : 'binary_crossentropy'},
        metrics = {
        'dept_output' : 'accuracy',
        'attr_output' : 'accuracy'}
        )

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled,
          {'dept_output' : dept_y_train_enc, 'attr_output' : attr_y_train_enc},
          epochs=100,
          batch_size=32,
          validation_split=0.2)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {"dept_output" : dept_y_test_enc, "attr_output" : attr_y_test_enc})

print(test_results)
```

```python
print(f"Department Accuracy: {test_results[2]} \nAttrition Accuracy: {test_results[1]}")
```
